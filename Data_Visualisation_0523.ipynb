```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
tips['tip_fraction']= tips['tip']/tips['total_bill']
tips.head()
```

```python
sns.scatterplot(x='total_bill',y='tip_fraction', data=tips)
```

```python
sns.scatterplot(x='total_bill',y='tip', data=tips, hue='time')
```

```python
sns.scatterplot(x='total_bill',y='tip', data=tips, hue='smoker')
```

```python
sns.scatterplot(x='total_bill',y='tip', data=tips, hue='time',style='sex', size='size');

```

```python
sns.scatterplot(x='total_bill',y='tip', data=tips, hue='time',style='sex', size='size');
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
sns.regplot(x='total_bill', y='tip',data=tips)
```

```python
sns.regplot(x='total_bill', y='tip_fraction', data=tips,marker='+')
```

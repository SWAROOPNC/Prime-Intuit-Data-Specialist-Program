```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,linewidth=1)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='smoker',linewidth=1)
```

```python
sns.set(color_codes = True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='smoker',linewidth=1,palette='Set2')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='smoker',linewidth=1,palette='Set2',dodge=True)
```

```python
sns.stripplot(x='time',y='tip',data=tips)
```

```python
sns.stripplot(x='time',y='tip',data=tips,order = ['Dinner','Lunch'])
```

```python
sns.stripplot(x='time',y='tip',data=tips,order = ['Dinner','Lunch'])
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import urllib.request
```

```python
# Getting the data file
# !wget https://api.covid19india.org/states_daily.json
#Unix way of getting the file

# In python, we use the url retrieve function

url = "https://api.covid19india.org/states_daily.json"

urllib.request.urlretrieve(url, 'data.json')
```

```python
covid_data = pd.read_json('data.json')
covid_data
```

```python
import json

with open('data.json') as f:
  data = json.load(f)
data
```

```python
data = data['states_daily']
data
```

```python
covid_data = pd.json_normalize(data)
covid_data
```

```python
covid_data.drop('dateymd', axis = 1, inplace = True)
cd = covid_data.tail(3)
```

```python

cd = covid_data.tail(3)
cd
```

```python
cd.set_index('status', inplace = True)
```

```python
cd.head()
```

```python
cd.drop('date', axis = 1, inplace = True )
```

```python
cd = cd.T
```

```python
cd.head()
```

```python
cd = cd.apply(pd.to_numeric)
cd 
```

```python
cd = cd.sort_values('Confirmed', ascending = False)
fig = plt.gcf()
fig
```

```python
fig.set_size_inches(15,6)

plt.bar(cd.index, cd.Confirmed, color = 'orange')
plt.bar(cd.index, cd.Recovered, bottom = cd.Confirmed, color = 'green')
plt.bar(cd.index, cd.Deceased, bottom = cd.Confirmed + cd.Recovered, color = 'red')

#plt.text(cd.index,cd.Confirmed, str(cd.Confirmed), ha = 'center')

for i, val in enumerate(cd.index): #for printing numbers
  y = cd.loc[val].sum() + 300
  if y > 3000:
    x = i
    plt.text(x,y,str(y-300), ha = 'center')
```

```python
cd = cd.sort_values('Confirmed', ascending = False)
fig = plt.gcf()
fig.set_size_inches(15,6)

plt.bar(cd.index, cd.Confirmed, color = 'orange')
plt.bar(cd.index, cd.Recovered, bottom = cd.Confirmed, color = 'green')
plt.bar(cd.index, cd.Deceased, bottom = cd.Confirmed + cd.Recovered, color = 'red')
```

```python
cd['Total'] = cd.sum(axis = 1)
cd['Confirmed_per'] = cd['Confirmed']/ cd['Total']
cd['Recovered_per'] = cd['Recovered']/ cd['Total']
cd['Deceased_per'] = cd['Deceased']/ cd['Total']
cd.head(5)
```

```python
cd = cd.sort_values('Total', ascending = False)
fig = plt.gcf()
fig.set_size_inches(15,6)
plt.bar(cd.index, cd.Confirmed_per, color = 'orange')
plt.bar(cd.index, cd.Recovered_per, bottom = cd.Confirmed_per, color = 'green')
plt.bar(cd.index, cd.Deceased_per, bottom = cd.Confirmed_per + cd.Recovered_per, color = 'red')
```

```python
cd = cd.sort_values('Total', ascending = False)
fig = plt.gcf()
fig.set_size_inches(10,10)

plt.barh(cd.index, cd.Confirmed_per, color = 'orange')
plt.barh(cd.index, cd.Recovered_per, left = cd.Confirmed_per, color = 'green')
plt.barh(cd.index, cd.Deceased_per, left = cd.Confirmed_per + cd.Recovered_per, color = 'red')
plt.yticks(rotation =90)
```

```python
def color_red_negative(x) :
    color = 'red' if x <0 else 'white'
    return 'color:' + color
```

```python
df2.style.applymap(color_red_negative) #applymap for rows # for each cells
```

```python
df2.style.highlight_max(color='red') #columnwise max
```

```python
df2.style.highlight_max(color='red',axis=1) #columnwise max
```

```python
df2.style.highlight_max(color='red').highlight_min(color='green')
```

```python
df2.drop('tt',axis=1,inplace=True)
df2
```

```python
def bold_max_value(x) :
    ismax=(x==x.max())
    return ['font-weight: bold' if y else '' for y in ismax]
```

```python
df2.style.apply(bold_max_value)
```

```python
df2.style.apply(bold_max_value).highlight_max(color='red') #statewise
```

```python
df2.style.apply(bold_max_value).highlight_max(color='red',axis=1) #rowwise , date wise ,max cases state
```
